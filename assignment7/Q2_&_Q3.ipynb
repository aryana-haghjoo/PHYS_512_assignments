I will use Lorantzians to generate exponensial deviations.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lorantzian Random number
def random_lorantzian(n):
  q= np.pi*(np.random.rand(n)-0.5)
  return np.tan(q)

#Loranzian distribution
def lorantzian_pdf(x):
  gamma=1
  return 1/(gamma* np.pi)* 1/(1+ np.power(x/gamma,2))

#exponensial distrribution
def exponensial(x):
  return 1/(np.sqrt(2*np.pi)) *np.exp(-0.5* x**2)
```

```python
n= 1000000
r_l = random_lorantzian(n)
r = np.random.rand(n)
M= 1.7

A= 1/(np.sqrt(2*np.pi))*np.exp(-0.5* r_l**2)
B= lorantzian_pdf(r_l)
keep = r < A/(M*B)
exp_rands = r_l[keep]
```

```python
plt.figure()
C = np.linspace(-10, 10, 10000)
plt.hist(r_l, bins=5, density=True)
plt.plot(C, exponensial(C), label='exp')
plt.plot(C, M * lorantzian_pdf(C), label='lorentz')
plt.legend()
plt.xlim(-10, 10)

plt.figure()
plt.hist(exp_rands, bins=30, density=True, label="Samples")
C= np.linspace(min(exp_rands), max(exp_rands), 1000)
plt.plot(C, 1/(np.sqrt(2*np.pi)) * np.exp(-0.5 * C**2), label='expected')
plt.legend()

print(f" Method is {len(exp_rands)/len(r_l) *100} % Efficient")
plt.show()
```

Chancing the M, will change our effiency because it makes the lorenzian come closer to the exponensial.
