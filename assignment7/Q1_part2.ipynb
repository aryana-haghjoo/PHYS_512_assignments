# The python random number generator.

```python
A = np.zeros((400,3))
for i in range (0,400):
    a= (1e8)* np.random.random_sample()
    b= (1e8)* np.random.random_sample()
    c= (1e8)* np.random.random_sample()
    A[i, 0] = a
    A[i, 1] = b
    A[i, 2] = c
```

```python
A[286,2]
```

```python
%matplotlib notebook

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# load some test data for demonstration and plot a wireframe
ax.scatter3D(A[:,0], A[:,1], A[:,2], cmap='Greens')


# rotate the axes and update
for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
```

As you can see, the python random number generator works really better than the C one.
