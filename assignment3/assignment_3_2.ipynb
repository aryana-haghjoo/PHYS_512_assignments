```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

table.JPG

```python
def fun(x,y, half_life = [1e21,1e10, 1e8, 5e16, 1e16, 3e14, 2e9, 1e6, 1e7, 8e6, 1, 5e12, 1e12, 7e10]):

  dydx=np.zeros(len(half_life)+1)

  dydx[0]= -y[0]/half_life[0] 

  for i in range(1,14):
    dydx[i]= y[i-1]/half_life[i-1] - y[i]/half_life[i]

  dydx[14]= y[13]/half_life[13]

  return(dydx)
```

```python
#initial conditions
y0= np.zeros(len(half_life)+1)
y0[0]= 1

t_span= [0,100]
t_eval=np.linspace(0,100,int(1e5))

A= integrate.solve_ivp(fun, t_span, y0,t_eval=t_eval)
B=A.y
```

```python
plt.plot(A.y[14,:]/A.y[0,:])
plt.ylabel('Pb206/U238')
plt.xlabel('time')
plt.show()
```
